###Selecting independent and dependent variable

Spliting Input column and the output column

```python
x=df[['wheelbase','carlength','carwidth','carheight','curbweight','enginesize','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg']]
```

```python
y=df['price']
```

```python
# Separating the data into tranning and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=20)
```

```python
#At first we train our sample data from dataset x_train and y_train
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
model=lr.fit(x_train,y_train)
```

```python
model_prediction=model.predict(x_train)
from sklearn.metrics import r2_score
print("Accuracy Score",r2_score(y_train,model_prediction)*100)
```

```python
model_test=lr.fit(x_test,y_test)
model_prediction_test=model.predict(x_test)
print("Accuracy Score",r2_score(y_test,model_prediction_test)*100)
```

```python
import numpy as np
x_test['Original Price']=y_test
x_test['Predicted Price']=np.round(model_prediction_test)
x_test.head()
```
