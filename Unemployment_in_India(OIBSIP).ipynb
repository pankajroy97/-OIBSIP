###The story behind this datasets is how lock-down affects employment opportunities and how the unemployment rate increases during the Covid-19.

This dataset contains the unemployment rate of all the states in India

Region = states in India

Date = date which the unemployment rate observed

Frequency = measuring frequency (Monthly)

Estimated Unemployment Rate (%) = percentage of people unemployed in each States of India

Estimated Employed = percentage of people employed

Estimated Labour Participation Rate (%) = labour force participation rate by dividing the number of people actively participating in the labour force by the
total number of people eligible to participate in the labor force

```python
#Data distribution based on state
plt.figure(figsize=(16,8))
sns.countplot(x=df['Region'])
plt.xticks(rotation=90)
plt.show()
```

```python
#Estimated unemployement rate based on Region
x=df.groupby('Region')[' Estimated Unemployment Rate (%)'].mean()
x1=x.sort_values(ascending=True)
print(x1)
plt.figure(figsize=(16,8))
sns.lineplot(x=x.index,y=x.values,marker='o')
plt.title('Region VS Avg Estimated Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
x=df.groupby('Region')[' Estimated Employed'].mean()
x1=x.sort_values(ascending=True)
print(x1)
plt.figure(figsize=(16,8))
sns.lineplot(x=x.index,y=x.values,marker='o',color='red')
plt.title('Region VS Avg Estimated Employed')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
x=df.groupby('Region')[' Estimated Labour Participation Rate (%)'].mean()
x1=x.sort_values(ascending=True)
print(x1)
plt.figure(figsize=(16,8))
sns.lineplot(x=x.index,y=x.values,marker='d',color='black')
plt.title('Region VS Avg Estimated Labour Participation Rate (%)')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```
